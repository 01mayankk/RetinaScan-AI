```python
# ============================================
# IMPORT REQUIRED LIBRARIES
# ============================================

import os
import random

import matplotlib.pyplot as plt

from PIL import Image
```

```python
# ============================================
# DATASET PATH
# ============================================

dataset_path = "../dataset/train"
```

```python
# ============================================
# DIABETIC RETINOPATHY CLASSES
# ============================================

classes = {

    "0": "No DR",

    "1": "Mild DR",

    "2": "Moderate DR",

    "3": "Severe DR",

    "4": "Proliferative DR"
}
```

```python
# ============================================
# DISPLAY SAMPLE RETINA IMAGES
# ============================================

plt.figure(figsize=(15, 10))

plot_index = 1

for class_id, class_name in classes.items():

    class_folder = os.path.join(
        dataset_path,
        class_id
    )

    # Get random image
    image_name = random.choice(
        os.listdir(class_folder)
    )

    image_path = os.path.join(
        class_folder,
        image_name
    )

    # Open image
    image = Image.open(image_path)

    # Plot image
    plt.subplot(2, 3, plot_index)

    plt.imshow(image)

    plt.title(f"{class_id} - {class_name}")

    plt.axis("off")

    plot_index += 1

plt.tight_layout()

plt.show()
```
